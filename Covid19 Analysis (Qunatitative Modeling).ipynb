```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("covid19.csv")
df.head()
```

```python
df.drop(["Sno"], axis = 1, inplace = True)
```

```python
df["State/UnionTerritory"].value_counts()
```

```python
df1 = df.groupby(['Date'], sort = False).sum()
```

```python
df1
```

```python
Total_cases = df1.sum(axis = 1)
df1['Total_cases'] = Total_cases 
df1
```

```python
df1.reset_index(inplace= True)
df1
```

```python
plt.figure(figsize=(20,5))
ax = sns.barplot(x ='Date',y= 'Total_cases' , data = df1 )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
```

```python
df2 = df1.loc[34:52]
df2
```

```python
df1.Date.astype("string")
df2["Date"] = pd.to_datetime(df2["Date"],format ='%d/%m/%y')

df2.Date.astype("string")
df2
```

```python
df2
```

```python

# r= (no of cases on next day- no of cases on current day)/(no of cases on current day) )

r = 0.0
start_date = df2.index[df2['Date']=='2020-03-04'][0]
end_date= df2.index[df2['Date']=='2020-03-21'][0]
for i in range(start_date, end_date):
    current_day = df2.loc[i]['Total_cases']
    next_day= df2.loc[i+1]['Total_cases']
    daily_rate=(next_day-current_day)/(current_day)
r = daily_rate
```

```python
r
```

```python
P_o=31
t=26
P_t = P_o*(np.exp(r*t))

P_t
```
